```python
# Dependencies
import pandas as pd
import numpy as np
import requests
import json
import pprint

# API Key
from config import api_key
```

```python
base_url = "https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol="
stocks_dict_list = {"Basic Industries" : ["PG", "UL", "UN", "BHP", "RIO"], 
               "Finance" : ["JPM", "BAC", "WFC", "HSBC", "C"],
               "Energy" : ["XOM", "CVX", "BP", "TOT", "BBL"],
               "Consumer Durables" : ["ABB", "KMB", "AME", "BLL", "CLX"],
               "Consumer Non-Durables" : ["KO", "PEP", "FMX", "BUD", "NKE"]}

keys = list(stocks_dict_list.keys())
values = list(stocks_dict_list.values())
```

```python
combined_stocks = values[0] + values[1] + values[2] + values[3] + values[4] 
volumes = []
closes = []
```

```python
for key in keys:
    five_stocks = stocks_dict_list[key]
    for stock in five_stocks:
        target_url = f"{base_url}{stock}&outputsize=full&apikey={api_key}"
        response = requests.get(target_url).json()
        try:
            rsp = response['Time Series (Daily)']
            dates = list(response['Time Series (Daily)'].keys())
        except KeyError:
            print("Error")
        for date in dates:
            closes.append(rsp[date]['4. close'])
            volumes.append(rsp[date]['5. volume'])
```

```python
# The volumes and closes columns are 25 times larger than the dates;
# this function splits these two columns into n equal sized chunks, same 
# length as the dates
```

```python
def chunks(l, n):
    """Yield successive n-sized chunks from l."""
    for i in range(0, len(l), n):
        yield l[i:i + n]
        
vols = list(chunks(volumes, len(dates)))

clos = list(chunks(closes, len(dates)))
```

```python
stock_dicts = {}
p = 0
for i in combined_stocks:
    stock_dict = {
        "Date" : dates,
        "Volume" : vols[p:p+1][0],
        "Close" : clos[p:p+1][0]
    }
  
    stock_dicts[combined_stocks[p]]=stock_dict
    p = p+1
```
